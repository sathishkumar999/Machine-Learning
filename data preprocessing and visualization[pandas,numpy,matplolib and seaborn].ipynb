```python
import matplotlib.pyplot as plt

plt.bar([1,3,5,7,9],[5,2,7,8,2], label="Example one",color='r')
plt.bar([2,4,6,8,10],[8,6,2,5,6], label="Example two", color='g')
plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')

plt.title('Epic Graph\nAnother Line! Whoa')

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[5,2,4,2,1,4,5,2]
plt.scatter(x,y,label='scatter',color='r',s=30,marker='*')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('scatter graph')
plt.legend()
plt.show()
```

```python
x=[30,20,25,25]
plt.pie(x)
plt.show()
```

```python
plt.pie(x,startangle=90)
plt.show()
```

```python
subjects=['maths','statistics','computers','physics']
plt.pie(x,labels=subjects,startangle=90)
plt.show()
```

```python
subjects=['maths','statistics','computers','physics']
plt.pie(x,labels=subjects,startangle=90,colors=['b','y','g','c'])
plt.show()
```

```python
subjects=['maths','statistics','computers','physics']
plt.pie(x,labels=subjects,startangle=90,colors=['b','y','g','c'],shadow=True,explode=(0.1,0,0,0),autopct='%1.1f%%')
plt.show()
```

# Seaborn for data visualizarion

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head()
```

```python
tips.columns
```

```python
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='sex',palette='winter_r')
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='smoker',palette='spring') #barplot
```

```python
sns.boxplot(x='day',y='tip',data=tips,palette='coolwarm') #boxplot
```

```python
sns.stripplot(x='day',y='tip',data=tips,palette='spring') #boxplot
```

```python
sns.stripplot(x='day',y='tip',data=tips,palette='spring',jitter=True) #stripplot
```

```python
sns.stripplot(x='sex',y='tip',data=tips,order=['Female','Male'],palette='spring') #stripplot
```

```python
#iris dataset for GridPlot
```

```python
iris=sns.load_dataset("iris")
```

```python
x=sns.PairGrid(iris)
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue='species',palette='coolwarm')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species',palette='spring')
x=x.map_diag(plt.hist,edgecolor='red')
x=x.map_offdiag(plt.scatter,edgecolor='green')
x=x.add_legend()
```
